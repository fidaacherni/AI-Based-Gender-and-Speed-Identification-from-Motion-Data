J'ai ajouter les colonnes des inputs avec la valeur NaN


```python
df
```

```python
L=[]
for i in range(1 ,1931):
    L.append("BDS0"+str(i).zfill(4)+".txt")
    
```

j'ai utilise la methodes zfill() pour entrer la forme des databases : "0000"--"1930"

```python
L
```

j'ai mis toutes les databases en une liste pour faciliter l'acees

```python
for i in range(len(L)):
    df_reader=pd.read_table(L[i])
    df.at[i,"Fx[N]"]=df_reader["Fx[N]"].mean()
    df.at[i,"Fy[N]"]=df_reader["Fy[N]"].mean()
    df.at[i,"Fz[N]"]=df_reader["Fz[N]"].mean()
    df.at[i,"Mx[Nm]"]=df_reader["Mx[Nm]"].mean()
    df.at[i,"My[Nm]"]=df_reader["My[Nm]"].mean()
    df.at[i,"Mz[Nm]"]=df_reader["Mz[Nm]"].mean()
    df.at[i,"COPx[cm]"]=df_reader["COPx[cm]"].mean()
    df.at[i,"COPy[cm]"]=df_reader["COPy[cm]"].mean()
    df.at[i,"Max_Fx"]=df_reader["Fx[N]"].max()
    df.at[i,"Max_Fy"]=df_reader["Fy[N]"].max()
    df.at[i,"Max_Fz"]=df_reader["Fz[N]"].max()
    df.at[i,"Min_Fx"]=df_reader["Fx[N]"].min()
    df.at[i,"Min_Fy"]=df_reader["Fy[N]"].min()
    df.at[i,"Min_Fz"]=df_reader["Fz[N]"].min()
    df.at[i,"Max_COPx"]=df_reader["COPx[cm]"].max()
    df.at[i,"Min_COPx"]=df_reader["COPx[cm]"].min()
    df.at[i,"Max_COPy"]=df_reader["COPy[cm]"].max()
    df.at[i,"Min_COPy"]=df_reader["COPy[cm]"].min()
    df.at[i,"Ecart_COPx"]=df_reader["COPx[cm]"].max()-df_reader["COPx[cm]"].min()
    df.at[i,"Ecart_COPy"]=df_reader["COPy[cm]"].max()-df_reader["COPy[cm]"].min()
    df.at[i,"Coeff_stab"]=math.sqrt((df_reader["Fx[N]"].mean()**2)+(df_reader["Fy[N]"].mean()**2))
    df.at[i,"Fx/Fz"]=(df_reader["Fx[N]"].mean()/df_reader["Fz[N]"].mean())
    df.at[i,"Fy/Fz"]=(df_reader["Fy[N]"].mean()/df_reader["Fz[N]"].mean())
    df.at[i,"Section"]=df.at[i,"Ecart_COPx"]*df.at[i,"Ecart_COPy"]
```
